# Business Understanding

## Project Overview

Organizations invest heavily in digital marketing campaigns to acquire and retain customers. A key question for marketing teams is whether these campaigns **actually influence customer behavior** or simply target customers who would have made a purchase regardless.

This project uses **causal inference** to estimate the **incremental impact** of a marketing campaign and identify which customers truly benefit from receiving the intervention.

---

# Business Problem

A retailer launches an email marketing campaign promoting new products to a subset of its customers.

The business wants to answer the following questions:

- Did the email campaign increase customer purchases?
- How much additional revenue was generated because of the campaign?
- Which customers should receive future marketing emails?
- How can the marketing budget be allocated more effectively?

These questions require understanding **causal impact**, not just customer behavior.

---

# What is Incrementality?

Incrementality measures the **additional outcome directly caused by a marketing intervention**.

For example, suppose 100 customers received an email and 20 made a purchase.

If 15 of those customers would have purchased even without the email, then the campaign only generated:

> **5 incremental purchases**

The objective is to isolate the effect of the campaign from customer behavior that would have occurred naturally.

---

# Why Prediction is Insufficient

Traditional machine learning focuses on predicting outcomes.

Typical predictive questions include:

- Which customers are likely to purchase?
- Which customers are likely to churn?
- Which customers have high lifetime value?

While these predictions are valuable, they do **not** answer whether an intervention changed customer behavior.

For example, a predictive model may identify customers who are already likely to purchase. Sending marketing emails to these customers may not increase sales because they would have purchased regardless.

Prediction answers:

> **Who will buy?**

Causal inference answers:

> **Who will buy because of the campaign?**

---

# Counterfactual Reasoning

The fundamental challenge in causal inference is that we never observe both outcomes for the same customer.

For every customer, there are two possible outcomes:

- Outcome if the customer receives the email
- Outcome if the customer does not receive the email

Only one of these outcomes is observed.

The unobserved outcome is called the **counterfactual**.

The goal of causal inference is to estimate this missing counterfactual so that we can measure the true effect of the treatment.

---

# Treatment and Control Groups

To estimate the effect of an intervention, customers are divided into two groups.

### Treatment Group

Customers who receive the marketing email.

### Control Group

Customers who do not receive the marketing email.

If both groups are comparable, differences in their outcomes can be attributed to the marketing campaign.

In observational datasets, however, treatment assignment is often not random, leading to selection bias. Causal inference methods help adjust for these differences before estimating treatment effects.

---

# Treatment Effects

## Average Treatment Effect (ATE)

The Average Treatment Effect measures the average impact of the marketing campaign across the entire customer population.

**Business Question**

> What is the average increase in purchases if every customer received the campaign?

ATE is useful for evaluating the overall effectiveness of a marketing strategy.

---

## Average Treatment Effect on the Treated (ATT)

ATT measures the effect of the campaign only for customers who actually received the treatment.

**Business Question**

> How much did the email increase purchases among customers who were targeted?

This metric is commonly used to evaluate existing marketing campaigns.

---

## Conditional Average Treatment Effect (CATE)

Not every customer responds to marketing in the same way.

CATE estimates treatment effects for different customer segments based on their characteristics.

Examples include:

- New vs. existing customers
- High-value vs. low-value customers
- Frequent vs. infrequent shoppers

Understanding heterogeneous treatment effects allows marketers to personalize campaigns and improve return on investment.

---

# Why Causal Inference Matters

Estimating causal effects enables organizations to move beyond prediction and make better business decisions.

Instead of targeting customers who are merely likely to purchase, companies can identify customers whose behavior is **most likely to change because of the intervention**.

This leads to:

- More effective marketing campaigns
- Better allocation of advertising budgets
- Improved customer targeting
- Higher return on investment
- Data-driven decision making

---

# Project Roadmap

This project will estimate advertising incrementality using several causal inference techniques.

1. Explore the dataset and understand customer characteristics.
2. Perform a naïve comparison between treatment and control groups.
3. Estimate propensity scores to adjust for selection bias.
4. Apply Propensity Score Matching (PSM).
5. Estimate treatment effects using Inverse Probability Weighting (IPW).
6. Implement Double Machine Learning (DML).
7. Build uplift models to identify customers with the highest incremental response.
8. Simulate marketing strategies and evaluate budget allocation policies.

The final objective is to estimate the true impact of the marketing campaign and provide actionable recommendations for future customer targeting.
